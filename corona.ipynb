```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/home/a/Downloads/covid.csv")
df.shape
```

```python
df.head()
```

```python
fill=df.fillna(0)
print(fill)
```

```python
fill.describe()
```

```python
continent_name=list(fill.continent.unique())
print(continent_name)
```

```python
group_a=fill.groupby(['date','location']) [['date','location','total_cases','total_deaths','total_vaccinations']].sum().reset_index()
group_a
```

```python
tot_det=group_a.loc[(group_a['total_deaths']>1000000)]
countries=tot_det['location'].unique()
abc=list(tot_det.location.unique())
abc
```

```python
for i in range(0,len(countries)):
    c=tot_det[tot_det['location']==countries[i]].reset_index()
    plt.scatter(np.arange(0,len(c)),c['total_cases'],color='green',label='total_cases')
    plt.scatter(np.arange(0,len(c)),c['total_deaths'],color='blue',label='total_deaths')
    plt.scatter(np.arange(0,len(c)),c['total_vaccinations'],color='red',label='total_vaccinations')
    plt.title(countries[i])
    plt.xlabel('no of days since first suspect')
    plt.ylabel('no of cases')
    plt.legend()
    plt.show()
```

```python
group_b=tot_det.groupby(['location']) [['location','total_cases','total_deaths']].sum().reset_index()
group_b
```

```python
c=group_b
plt.scatter(np.arange(0,len(c)),c['total_cases'],color='green',label='total_cases')
plt.scatter(np.arange(0,len(c)),c['total_deaths'],color='blue',label='total_deaths')
plt.title('world')
plt.xlabel('no of days since first suspect')
plt.ylabel('no of cases')
plt.legend()
plt.show()
```
